```python
import numpy as np

# Step 1: Define the activation function (step function)
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Step 2: Perceptron class
class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.zeros(input_size + 1)  # +1 for bias
        self.learning_rate = learning_rate
        self.epochs = epochs

    def predict(self, x):
        x = np.insert(x, 0, 1)  # Insert 1 for bias weight
        return step_function(np.dot(x, self.weights))

    def train(self, X, y):
        for _ in range(self.epochs):
            for i in range(len(X)):
                x_i = np.insert(X[i], 0, 1)  # Insert 1 for bias
                prediction = self.predict(X[i])
                error = y[i] - prediction
                # Update weights
                self.weights += self.learning_rate * error * x_i

# Example usage
# Define input data (X) and target labels (y)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])  # AND logic gate

# Create Perceptron instance and train it
perceptron = Perceptron(input_size=2)
perceptron.train(X, y)

# Test the perceptron
for x in X:
    print(f"Input: {x}, Predicted: {perceptron.predict(x)}")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

# Step 1: Prepare the dataset (AND gate example)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])  # AND gate output

# Step 2: Define the model
model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))  # Single-layer perceptron

# Step 3: Compile the model
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])

# Step 4: Train the model
model.fit(X, y, epochs=100, verbose=1)

# Step 5: Evaluate and test the model
loss, accuracy = model.evaluate(X, y)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Predicting on the training data
predictions = model.predict(X)
for i, x in enumerate(X):
    print(f"Input: {x}, Predicted: {predictions[i][0]:.2f}, Actual: {y[i]}")
```
